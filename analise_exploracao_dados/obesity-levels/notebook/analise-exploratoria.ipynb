# Configuração para os gráficos

```python
# definicao de cores
GRAY1, GRAY2, GRAY3 = '#231F20', '#414040', '#555655'
GRAY4, GRAY5, GRAY6 = '#646369', '#76787B', '#828282'
GRAY7, GRAY8, GRAY9 = '#929497', '#A6A6A5', '#BFBEBE'
BLUE1, BLUE2, BLUE3, BLUE4 = '#174A7E', '#4A81BF', '#94B2D7', '#94AFC5'
RED1, RED2 = '#C3514E', '#E6BAB7'
GREEN1, GREEN2 = '#0C8040', '#9ABB59'
ORANGE1 = '#F79747'

# configura a família de fontes do plot para Arial
plt.rcParams['font.family'] = 'Arial'

# configura a família de fontes mathtext em negrito e itálico para Arial
matplotlib.rcParams['mathtext.fontset'] = 'custom'
matplotlib.rcParams['mathtext.bf'] = 'Arial:bold'
matplotlib.rcParams['mathtext.it'] = 'Arial:italic'
```

```python
class Grafico:
    def __init__(self, config_padrao):
        self.__config_grafico = config_padrao

        self.__figura, self.__graficos = plt.subplots(self.__config_grafico['figura']['linhas'], 
                                                      self.__config_grafico['figura']['colunas'], 
                                                      figsize=self.__config_grafico['figura']['tamanho'], 
                                                      width_ratios=self.__config_grafico['figura'].get('proporcao', None), 
                                                      tight_layout=True)
        
        # Título e subtítulo do gráfico
        self.__figura.suptitle(self.__config_grafico['titulo'], x=0.01, y=1, horizontalalignment='left', verticalalignment='top', fontsize = 25, color=GRAY4, fontweight='bold')

        if self.__config_grafico['subtitulo'] is not None:
            self.__figura.text(0.01, 0.92, self.__config_grafico['subtitulo'], horizontalalignment='left', verticalalignment='top', fontstyle="italic",  fontsize=16, color=GRAY7, fontweight='bold')
        
        sns.despine(left=True)

    def grafico_barras_relativo(self, config_grafico, numero_grafico=0):
        total_graficos = self.__config_grafico['figura']['linhas'] * self.__config_grafico['figura']['colunas']
        ax = self.__graficos if total_graficos == 1 else self.__graficos[numero_grafico]
            
        ax = sns.barplot(data=config_grafico['data'], x=config_grafico['x'], y=config_grafico['y'], hue=config_grafico.get('hue', None), palette=config_grafico['palette'], ax=ax)
        ax.set_title(config_grafico['title'], fontsize=13, color=GRAY3)
        ax.set(xlabel=config_grafico['xlabel'], ylabel=config_grafico['ylabel'])
        ax.set_xticklabels(ax.get_xticklabels(), rotation=config_grafico['rotacao_texto_x'])
       
        if all(k in config_grafico for k in ['coord_anotacao_x', 'coord_anotacao_y', 'anotacao']):
            ax.text(
                config_grafico['coord_anotacao_x'], 
                config_grafico['coord_anotacao_y'], 
                config_grafico['anotacao'], 
                ha="left", 
                va="center", 
                size=10
            )

        if config_grafico['mostrar_eixo_y'] is not None:
            ax.get_yaxis().set_ticks([])
        
        # Adiciona rótulos nas barras
        for c in ax.containers:
            if config_grafico['simbolo_porc'] == True:
                ax.bar_label(c, fmt=lambda x: f'{x:0.0f}%' if x > 0 else '', padding=config_grafico['padding'], color=config_grafico['color'], fontweight='bold')
            else:
                ax.bar_label(c, fmt=lambda x: f'{x:0.0f}' if x > 0 else '', padding=config_grafico['padding'], color=config_grafico['color'], fontweight='bold')
        
    def grafico_histograma(self, config_grafico, numero_grafico):
        sns.histplot(data=config_grafico['data'], x=config_grafico['x'], color=config_grafico['color'], ax=self.__graficos[numero_grafico], kde=True)
        sns.despine()

    def grafico_boxplot(self, config_grafico, numero_grafico):
        sns.boxplot(data=config_grafico['data'], x=config_grafico['x'], color=config_grafico['color'], ax=self.__graficos[numero_grafico])
        sns.despine()
```

Como o objetivo do estudo é investigar os principais fatores associados à obesidade, não iremos entrar a fundo na analise entre os graus de obesidade e sobrepeso. Com isso fizemos uma coluna no dataset para ter apenas Obesidade, sobrepeso, Normal e Insuficiente (peso insuficiente). Aqui iremos criar uma coluna chamada faixa-etaria para ver a distribuição das idades, facilitando as análises.

```python
# Convertendo os valores da coluna de idade para inteiros
df_obesity['Age'] = df_obesity['Age'].astype(int)

# Determinando o mínimo e o máximo valor da coluna de idade
idade_min = df_obesity['Age'].min()
idade_max = df_obesity['Age'].max()

# Determinando as faixas etárias
faixas_etarias = [(i, i + 4) for i in range(idade_min, idade_max + 1, 5)]
rotulos = [f"{faixa[0]}-{faixa[1]}" for faixa in faixas_etarias]

# Função para atribuir a faixa etária com base na idade
def atribuir_faixa_etaria(idade):
    for faixa, rotulo in zip(faixas_etarias, rotulos):
        if faixa[0] <= idade <= faixa[1]:
            return rotulo

# Aplicando a função para criar a coluna de faixa etária
df_obesity['Faixa_Etaria'] = df_obesity['Age'].apply(atribuir_faixa_etaria)
df_obesity
```

## Impacto do tempo de utilização de dispositivos eletrônicos: 

- TUE: Recurso, Inteiro, “Quanto tempo você utiliza dispositivos tecnológicos como celular, videogame, televisão, computador e outros?”

```python
df_obesity.TUE
```

```python
df_obesity.TUE.describe()
```

Vamos assumir para essa análise que o tempo de utilização dos dispositivos eletrônicos é em horas variando do mínimo ao máximo observado na coluna. E também é em relação ao dia de uso.

```python
# Função para classificar os valores
def classificar_tue(valor):
    if valor == 0:
        return '0'
    elif valor == 1:
        return '1'
    elif valor > 1 and valor < 2:
        return '1 a 2'
    else:
        return 'Mais de 2'

# Aplicando a função para criar a nova coluna
df_obesity['TUE_Classificacao'] = df_obesity['TUE'].apply(classificar_tue)
```

```python
colors = [BLUE1, BLUE2, BLUE3, BLUE4]

config_regiao = {'figura': {'linhas': 1, 'colunas': 2, 'proporcao': [1,2], 'tamanho': (14,7)},
                'titulo': "Meio de Transporte utilizado pelos entrevistados\n\n",
                'subtitulo': None}

grafico_1 = {
    'data': df_obesity.TUE_Classificacao.value_counts(normalize=True).mul(100).round(2).rename('porcentagem').reset_index(),
    'x': 'TUE_Classificacao',
    'y': 'porcentagem',
    'title': '45% dos entrevistados usam dispositivos tecnológicos\n 2 ou mais horas no dia',
    'xlabel': '',
    'ylabel': 'Quantidade relativa (% da TUE)',
    'rotacao_texto_x': 45,
    'palette': colors,
    'padding':3,
    'color':GRAY3,
    'simbolo_porc':True,
    'mostrar_eixo_y':True
}

grafico_2 = {
    'data': df_obesity.groupby('TUE_Classificacao')['categoria_obesity'].value_counts(normalize=True).mul(100).round(2).rename('porcentagem').reset_index(),
    'x': 'porcentagem',
    'y': 'TUE_Classificacao',
    'hue': 'categoria_obesity',
    'title': 'Distribuição da classificação dos pesos dos entrevistados',
    'xlabel': 'Quantidade relativa (% categoria de peso)',
    'ylabel': '',
    'rotacao_texto_x': 0,
    'palette': colors,
    'padding':3,
    'color':GRAY3,
    'simbolo_porc':True,
    'mostrar_eixo_y':None
}


grafico_regiao = Grafico(config_regiao)
grafico_regiao.grafico_barras_relativo(grafico_1, 0)
grafico_regiao.grafico_barras_relativo(grafico_2, 1)
```

Note-se que:
